```python
class Students:
    
    def __init__(self,name,mssv,ddy,address,majors,cls):
        self.name=name
        self.mssv=mssv
        self.ddy=ddy
        self.address=address
        self.majors=majors
        self.cls=cls
    def show(self):
        print("Họ và tên: %s, MSSV: %s, DDY: %s, Địa chỉ: %s, Ngành: %s, Lớp:%s"\
              %(self.name,self.mssv,self.ddy,self.address,self.majors,self.cls))
def accept(lst):
    name=input("Nhập tên: ")
    mssv=input("Nhập MSSV: ")
    ddy=input("Nhập ngày tháng năm sinh: ")
    address=input("Nhập địa chỉ: ")
    majors=input("Nhập ngành học: ")
    cls=input("Nhập lớp học: ")    
    student=Students(name,mssv,ddy,address,majors,cls)
    lst.append(student)
    
def show_lst(lst):
    for i in range(len(lst)):
        lst[i].show()
        
def find_sv(lst,name):
    check=False
    for i in range(len(lst)):
        if lst[i].name==name:
            lst[i].show()
            check=True
    if check==False:
        print("Sinh viên không có trong danh sách")

def del_sv(lst,mssv):
    check=False
    for i in range(len(lst)):
        if lst[i].mssv==mssv:
            check=True
            print("Bạn đang xóa SV", lst[i].name)
            while True:
                choice=input("Bạn có muồn tiếp tục không? (y/n): ")
                if choice.lower() in ['y','n']:
                    break
                else:
                    print("Bạn nhập sai cú pháp!")
            if choice.lower() == 'y':
                del lst[i]
                print("Đã xóa")
    if check==False:
        print("Sinh viên không có trong danh sách")
        
def adjust_sv(lst,name):
    check=False
    for i in range(len(lst)):
        if lst[i].name==name:
            print("Bạn đang chỉnh sửa thông tin sinh viên", lst[i].name)
            student=lst[i]
            check=True
            break
    if check == False:
        print("Sinh viên không có trong danh sách")
    else:
        while True: 
            choice=input("Bạn có muồn tiếp tục không? (y/n): ")
            if choice.lower() in ['y','n']:
                break
            else:
                print("'Bạn nhập sai cú pháp!")
        if choice=='y':
            name=input("Nhập tên: ")
            mssv=input("Nhập MSSV: ")
            ddy=input("Nhập ngày tháng năm sinh: ")
            address=input("Nhập địa chỉ: ")
            majors=input("Nhập ngành học: ")
            cls=input("Nhập lớp học: ")  

def sx_lst_mssv(lst):
    sorted(lst,key=lambda x:x.mssv)
    
def con():
    global run
    while True:
        choice=input("Bạn có muốn tiếp tục không? (y/n)")
        if choice.lower() == 'n':
            run=False
            break
        elif choice.lower()=='y':
            break
        else:
            print("Bạn nhập sai cú pháp")
            


lst_student=[]
run=True
while run:
    print('Option: \n\
    1. Thêm SV vào danh sách \n\
    2. Tìm kiếm SV họ và tên \n\
    3. Xóa SV khỏi danh sách theo MSSV\n\
    4. Cập nhật dữ liệu SV theo tên\n\
    5. Xuất danh sách SV\n\
    6. Xuất danh sách theo MSSV\n\
    7. Thoát')
    choice=input("Lựa chọn của bạn: ")
    if choice == '1':
        accept(lst_student)
        con()
    elif choice == '2':
        name=input("Nhập tên hoặc MSSV của SV mà bạn muốn tìm: ")
        find_sv(lst_student,name)
        con()
    elif choice == '3':
        name=input("Nhập MSSV của SV mà bạn muốn xóa: ")
        del_sv(lst_student,name)
        con()
    elif choice == '4':
        name=input("Nhập MSSV của SV mà bạn muốn chỉnh sửa: ")
        adjust_sv(lst_student,name)
        con()
    elif choice == '5':
        show_lst(lst_student)
    elif choice == '6':
        sx_lst_mssv(lst_student)
        show_lst(lst_student)
        con()
    elif choice == '7':
        run= False
    else:
        print("Bạn nhập sai cú pháp")
```
